```python
import pymongo
from tabulate import tabulate
import pandas as pd

# Connect to MongoDB
client = pymongo.MongoClient("mongodb://localhost:27017/")

def find_query(dbName, collectionName, query, projection, sort_col=None, limit=None):
    db = client[dbName]  # Specify the database name
    collection = db[collectionName]  # Specify the collection name
    
    # Construct and print the query in MongoDB shell format and its English description
    query_str = f"db.{collectionName}.find({query}, {projection})"
    if sort_col:
        query_str += f".sort({sort_col})"
    if limit:
        query_str += f".limit({limit})"
    print("Nosql Query:")
    print(query_str)
    print("\nEnglish Counterpart:")
    print("This query finds all accidents where the number of casualties is between 30 and 100 and it lists accident's index, severity, date, road surface conditions, and weather conditions for each accident.\n")
    
    # Execute the query
    cursor = collection.find(query, projection)
    if sort_col:
        cursor = cursor.sort(list(sort_col.items()))
    if limit:
        cursor = cursor.limit(limit)
    documents = list(cursor)
    if documents:
        df = pd.DataFrame(documents)
        column_names = df.columns.tolist()
        result_table = tabulate(df, headers=column_names, tablefmt="grid", showindex=False)
        print("Query results:")
        print(result_table)
    else:
        print("No documents found")
        
        
        
def aggregate_query(dbName, collectionName, pipeline):
    db = client[dbName]  # Specify the database name
    collection = db[collectionName]  # Specify the collection name
    print("Executing Aggregate Pipeline:")
    print(f"db.{collectionName}.aggregate({pipeline})")
    print("\nDescription:")
    print(f"This pipeline aggregates documents in the '{collectionName}' collection, grouping them by branch number, and computes the count and sum of salaries per branch.\n")
    
    pipe_output = collection.aggregate(pipeline)
    first_document = next(pipe_output, None)  # Get the first document or None if no documents
    if first_document:
        column_names = list(first_document.keys())
        documents = list(pipe_output)
        df = pd.DataFrame(documents)
        result_table = tabulate(df, headers=column_names, tablefmt="grid", showindex=False)
        print("Query results:")
        print(result_table)
    else:
        print("No documents found")
        


# Define the specific MongoDB find() query parameters for the 'Accidents' collection
db_name = 'Project_DW'  # Replace with your actual database name
collection_name = 'Accidents'
query = {
    '$and': [
        {'Number_of_Casualties': {'$gte': 30}},
        {'Number_of_Casualties': {'$lte': 100}}
    ]
}
projection = {
    "_id": 0,
    "Accident_Index": 1,  
    "Accident_Severity": 1,
    "Date": 1,
    "Road_Surface_Conditions": 1,
    "Weather_Conditions": 1
}

# Execute the find_query function for 'Accidents' collection
find_query(db_name, collection_name, query, projection)
```
